```python
#Este es el link en donde están los datos:
url = "https://apps.mades.gov.py/siam/portal/disponibilidad"
url2 = "https://apps.mades.gov.py/siam/portal/areas-certificadas"
url3= "https://apps.mades.gov.py/siam/portal/transaccion"
```

```python
url = "https://apps.mades.gov.py/siam/portal/transaccion"


rows = []

for start in range(0, 2200, 10):
    params = {
        "iDisplayStart": start,
        "iDisplayLength": 10
    }

    r = requests.get(url, params=params)
    data = r.json()

    # aaData es una lista de diccionarios
    rows.extend(data["aaData"])

df = pd.DataFrame(rows)
df
```
